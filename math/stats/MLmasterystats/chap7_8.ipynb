# Law of Large Numbers

The `law of large numbers` is a theorem stating that the average result from repeating an experiment multiple times will better approximate the expected result

In other words, as we increase our **sample size**, the **sample mean** will better approximate the expected **population mean**.

In `machine learning lingo`: the more training data we have, **generally** the better model


```python
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
sns.set_style("ticks")

np.random.seed(1)
# create sample sizes
sizes = [x for x in range (10, 20000, 200)]

# generate samples ofdifferent siaes and calculate their means
# mean = 50, std=5
means = [np.mean(5 * randn(size) + 50) for size in sizes]


sns.scatterplot(x=sizes, y=np.array(means)-50)
plt.xlabel("Sample sizes")
plt.ylabel("Error")
plt.show()
```

# Central Limit Theorem (CLT)

The `central limit theorem` states that as the size of the sample increases, the **distribution of the means** across the multiple samples will approximate a **Gaussian distribution**

In other words, if we draw multiple independent samples and calculate their means, the distribution of those means will form a Gaussian curve

Interestingly, this will occur `no matter the shape of the population` we are drawing samples from

The `CLT` and `Law of Large Numbers (LLN)` are similar, but ultimately different:
- `LLN` - as the size of a sample is increased, the more accurate of an estimate the sample mean will be of the population mean
- `CLT` - the means of multiple independent samples will form a Gaussian distribution itself.

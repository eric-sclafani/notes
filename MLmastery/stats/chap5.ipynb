## Line Plots

`Line plots` are used to present observations at regular intervals (such as time).

The **x-axis** represents the regular interval

The **y-axis** shows the observations ordered by the x-axis and connected by a line.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np

sns.set_theme()

# interval for the x-axis (multiplying by 0.1 here just scales the values down)
x = [i*0.1 for i in range (100)]

# function of x for the y-axis
y = np.sin(x)

plt.xlabel("Regular interval")
plt.ylabel("Observations")
sns.lineplot(x=x, y=y)
plt.show()
```

## Bar charts

`Bar charts` are used to present relative quantities over multiple categories.

The **x-axis** represents the categories and are spaced evenly

The **y-axis** represents the quantity for each category and is drawn as a bar from the baseline to the appropriate level on the y-axis

`Bar charts` are useful for showing the categorical distribution of your data

```python
from random import randint

# x-axis categories
x = ["Red", "Blue", "Green"]

# quantities for each category (random values)
y = [randint(0,100), randint(0,100), randint(0,100)]

plt.xlabel("Categories")
plt.ylabel("Quantities")
sns.barplot(x=x, y=y)
plt.show()
```

## Histograms

`Histograms` are used to show the distribution of your data (i.e. gaussian, uniform, long tail, etc...). 

The **x-axis** represents discrete bins or intervals for the observations. 

The **y-axis** represents the frequency of observations that belong to each bin

```python
# pull random numbers from a normal distribution
x = np.random.randn(1000)

fig, axes = plt.subplots(1,2, figsize=(10,5))

sns.histplot(x, bins=10, color="green", ax=axes[0], kde=True)
sns.histplot(x, bins=100, color="red", ax=axes[1], kde=True)

axes[0].title.set_text("10 bins")
axes[1].title.set_text("100 bins")

plt.show()
```

## Box plots

`Box plots` are also used to show the distibution of data. This method is often used for non-Gaussian data. `Box plots` are better at showing things like **location, skewness, spread, tile length, and outlying points**

The **x-axis** represents your data samples

The **y-axis** represents the observation values

Notes about boxplots:

- A `box` is drawn to summarize the middle 50% of the dataset *starting* at the observation at the **25th percentile** and *ending* at the **75th percentile**
    - `25th percentile` = **lower quartile**
    - `75th percentile` = **Upper quartile**

- The `median`, or 50th percentile is drawn with a **line**

- A value called the `Interquartile range` or `IQR` is calculated as $1.5*$ the **difference** between the 75th and 25th percentiles.

- Lines called `whiskers` are drawn extending from both ends of the box with the length of the `IQR` to demonstrate the range of sensible values in the distribution

- Observations outside the whiskers might be outliers and drawn with `small circles`

```python

# random numbers pulled from normal distribution
x = [np.random.randn(1000), 5*np.random.randn(1000), 10*np.random.randn(1000)]

plt.boxplot(x)
plt.show()
```

## Scatterplots

`Scatterplots` are used the summarize the relationship between two paired data samples. 

**Paired** data samples means that two measures were recorded for a given observation, such as height and weight of a person

A **correlation** can be quantified between the two variables, such as a `line of best fit`

The **x-axis** represents the value of the first variable

The **y-axis** represents the value of the second variable

```python
x = 20 * np.random.randn(1000) + 100

y = x + (10 * np.random.randn(1000) + 50)

sns.regplot(x=x, y=y)
#sns.scatterplot(x=x, y=y)
plt.show()
```

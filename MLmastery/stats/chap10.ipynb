# Probability distribution functions (PDFs)


We can think of a `distribution` as a **function** that *describes the relationship between observations* in a sample space

A distribution is simply a **collection of data** on a variable. Usually, these scores are *arranged from smallest to largest*

- `Discrete variables` - Discrete variables refers to _**countable, individualized items**_. These items are not divisible. Instead, they only exist in set increments or units. 

    - Basically, theres a `fixed range of numbers` that a variable can be drawn from.
    - For instance, the **number of students in a classroom** is a discrete value.  
    - Another example is dice. With a normal die, there are only ever 6 possible outcomes, not 6.5, 5.5.
    
&nbsp;
- `Continuous variables` - Continuous variables is **data that is measurable**, versus being countable. Continuous variables use increments that are dividable and subcountable. In other words, continuous variables could go on forever.

    - People's heights, weights
    - Product dimensions
    - Time spent on a website

### Probability Density Functions (PDF)

`Probability density function` is used to calculate the likelihood of an observation in a  distribution. PDFs operate over **`continuous variables`**.

The `PDF` can be used to summaraize the **likelihood of observations** across the distribution's sample space.

### Cumulative Density Function (CDF)

`Cumulative density function` (also referred to as `cumulative distribution function`) calculates the cumulative likelihood for an observation \
and all prior observations in the sample space.

It allows us to quickly understand **how much of the distribution** lies `before and after a given value`.

The `CDF` is plotted as a curve from 0 to 1.


### Probability Mass Functions (PMF)

`Probability mass function` is used to describe a distribution of **`discrete`** variables.

 Can be thought of as the equivalent to the PDF that operates over continuous variables instead of discrete.


 ----

 Using `scipy`, we can create a gaussian probability density function with a **given sample space, mean, and std**

 We can also call scipy's `norm.cdf` function to create a gaussian cumulative density function 



```python
from scipy.stats import norm

# define our sample space (-5 to 5, step=0.001)
sample_space = np.arange(-5,5,0.001)
mean = 0.0
std = 1.0

pdf = norm.pdf(sample_space, mean, std)

sns.lineplot(x=sample_space, y=pdf, color="lightblue")
plt.xlabel("Sample space")
plt.ylabel("Likelihood of each")
plt.show()
```

```python
cdf = norm.cdf(sample_space)

sns.lineplot(x=sample_space, y=cdf, color="orange")
plt.xlabel("Sample space")
plt.ylabel("Cumulative probability")
plt.show()
```

Running the example above creates an `s curve`. 

From the graph, we can see that a **value of 2 covers close to 100% of the graph**, with only a veyr small thin tail beyond that point.

We can also see that the **mean value of zero** shows 50% of the observations before and after that point.
